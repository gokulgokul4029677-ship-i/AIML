```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = {
    "Restaurant": [
        "Spice Hub", "Pasta Palace", "Sushi World", "Burger Barn", "Curry Corner",
        "Taco Town", "Dragon Wok", "Pizza Plaza", "BBQ House", "Vegan Delight"
    ],
    "Cuisine": [
        "Indian", "Italian", "Japanese", "American", "Indian",
        "Mexican", "Chinese", "Italian", "American", "Vegan"
    ],
    "Price_Range": [
        "Medium", "High", "High", "Low", "Medium",
        "Low", "Medium", "Medium", "High", "Medium"
    ],
    "Rating": [
        4.5, 4.2, 4.8, 4.0, 4.3,
        4.1, 4.4, 4.0, 4.6, 4.7
    ]
}

df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df["Cuisine"].fillna("Unknown", inplace=True)
df["Price_Range"].fillna("Medium", inplace=True)
df["Rating"].fillna(df["Rating"].mean(), inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

encoded_features = encoder.fit_transform(df[["Cuisine", "Price_Range"]])

encoded_df = pd.DataFrame(
    encoded_features,
    columns=encoder.get_feature_names_out(["Cuisine", "Price_Range"])
)

encoded_df
```

```python
final_df = pd.concat([encoded_df, df["Rating"]], axis=1)
final_df
```

```python
similarity_matrix = cosine_similarity(final_df)
```

```python
def recommend_restaurants(restaurant_name, top_n=3):
    index = df[df["Restaurant"] == restaurant_name].index[0]
    similarity_scores = list(enumerate(similarity_matrix[index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    recommended_indices = [i[0] for i in similarity_scores[1:top_n+1]]
    return df.iloc[recommended_indices][["Restaurant", "Cuisine", "Price_Range", "Rating"]]
```

```python
recommend_restaurants("Spice Hub")
```

```python
user_preference = {
    "Cuisine": "Indian",
    "Price_Range": "Medium"
}

user_df = pd.DataFrame([user_preference])
user_encoded = encoder.transform(user_df)

user_vector = np.concatenate([user_encoded[0], [df["Rating"].mean()]])
```

```python
user_similarity = cosine_similarity([user_vector], final_df)
top_indices = user_similarity[0].argsort()[::-1][:3]

df.iloc[top_indices][["Restaurant", "Cuisine", "Price_Range", "Rating"]]
```
